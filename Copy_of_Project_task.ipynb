```python
import numpy as np

products = ["Laptop", "Mobile", "Tablet", "Headphone", "Smartwatch"]

sales = np.array([120, 250, 80, 250, 100])

print("Total Sales:", np.sum(sales))
print("Average Sales:", np.mean(sales))
print("Highest Sales:", np.max(sales))
print("Lowest Sales:", np.min(sales))

best_index = np.argmax(sales)
print("Best Selling Product:", products[best_index])

print("\nProduct Sales Report")
for i in range(len(products)):
    print(products[i], ":", sales[i])
```
