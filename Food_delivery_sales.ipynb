```python
import pandas as pd

data = {
    'Food Item': ['Biryani', 'Pizza', 'Burger', 'Dosa', 'Burger', 'Biryani'],
    'Order Time': ['13:00', '20:00', '19:00', '09:00', '18:00', '13:00'],
    'Location': ['Madurai', 'Chennai', 'Madurai', 'Madurai', 'Trichy', 'Chennai'],
    'Sales': [450, 600, 300, 500, 650, 200]
}

df = pd.DataFrame(data)
df.head()
```

```python
cleaned_df = df.dropna()
cleaned_df.head()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt

food_count = df['Food Item'].value_counts()

food_count.plot(kind='bar')
plt.title("Most Popular Food Items")
plt.xlabel("Food Item")
plt.ylabel("Orders")
plt.show()
```

```python
location_sales = df.groupby('Location')['Sales'].sum()

location_sales.plot(kind='bar')
plt.title("Location Wise Sales")
plt.xlabel("Location")
plt.ylabel("Total Sales")
plt.show()
```

```python
time_count = df['Order Time'].value_counts()

time_count.plot(kind='line', marker='o')
plt.title("Peak Order Time Analysis")
plt.xlabel("Order Time")
plt.ylabel("Orders")
plt.show()
```
